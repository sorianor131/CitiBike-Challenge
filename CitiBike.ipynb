```python
import pandas as pd
```

```python
# Read in CSV files for each month
MarDF = pd.read_csv('Resources/JC-202003-citibike-tripdata.csv')
AprDF = pd.read_csv('Resources/JC-202004-citibike-tripdata.csv')
MayDF = pd.read_csv('Resources/JC-202005-citibike-tripdata.csv')
JunDF = pd.read_csv('Resources/JC-202006-citibike-tripdata.csv')
JulDF = pd.read_csv('Resources/JC-202007-citibike-tripdata.csv')
AugDF = pd.read_csv('Resources/JC-202008-citibike-tripdata.csv')
SepDF = pd.read_csv('Resources/JC-202009-citibike-tripdata.csv')
OctDF = pd.read_csv('Resources/JC-202010-citibike-tripdata.csv')
NovDF = pd.read_csv('Resources/JC-202011-citibike-tripdata.csv')
DecDF = pd.read_csv('Resources/JC-202012-citibike-tripdata.csv')
```
